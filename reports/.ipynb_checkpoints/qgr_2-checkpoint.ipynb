```python
def volume(df, routine, volume_type):
    '''
    Create pivot table of total volume or relative volume
    per muscle group each week, sorting by the most recent 
    week's volume
    '''
    view = (df
            .loc[df.routine == routine]
            .pivot_table(index = 'date', columns = 'category', values = volume_type, aggfunc = 'sum')
            .resample('W')
            .sum()
            .astype(int)
            .drop('Cardio', axis = 1))
    
    sorted_view = view.sort_values(by = view.index[-1], axis = 1, ascending = False)
        
    
    return sorted_view
```

```python
vol = volume(sets, 'Episode VI: Return of the Legs', 'volume')
```
